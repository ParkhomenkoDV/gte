```python
for method, value in METHODS.items():
    print(method)
    for k, v in value.items():
        if k == 'attributes':
            for k_, v_ in v.items():
                print(f'\t{k_}: {v_}')
        else:
            print(f'{k}: {v}')
    print()
```

```python
figsize: tuple = (15, 15)  # размер изображения в дюймах
```

```python
for foil in (*stator_foils, *rotor_foils):
    fig = foil.plot(figsize=figsize)
    fig.savefig(f'./exports/{foil.name}.png')
    fig.show()
    foil.write('xlsx', folder='exports', relative=False, header=False)
```
