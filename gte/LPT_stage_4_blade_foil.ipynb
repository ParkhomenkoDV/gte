```python
for method, value in METHODS.items():
    print(method)
    for k, v in value.items():
        if k == 'attributes':
            for k_, v_ in v.items():
                print(f'\t{k_}: {v_}')
        else:
            print(f'{k}: {v}')
    print()
```

```python
discreteness: int = 20
```

```python
stator_foils = (
    Foil('BMSTU', discreteness, 
         chord=69.4, installation_angle=radians(136.4-90), step=43.3, name='stator_1',
         rotation_angle=radians(51.72),
         relative_inlet_radius=0.028, relative_outlet_radius=0.014,
         inlet_angle=radians(10), outlet_angle=radians(10),
         x_ray_cross=0.4,
         upper_proximity=0.5),
    Foil('BMSTU', discreteness, 
         chord=69.4, installation_angle=radians(136.6-90), step=46.6, name='stator_2',
         rotation_angle=radians(57.46),
         relative_inlet_radius=0.036, relative_outlet_radius=0.018,
         inlet_angle=radians(15), outlet_angle=radians(10),
         x_ray_cross=0.4,
         upper_proximity=0.5),
    Foil('BMSTU', discreteness, 
         chord=69.4, installation_angle=radians(136.2-90), step=49.8, name='stator_3',
         rotation_angle=radians(62.72),
         relative_inlet_radius=0.044, relative_outlet_radius=0.022,
         inlet_angle=radians(20), outlet_angle=radians(10),
         x_ray_cross=0.4,
         upper_proximity=0.5),
)

rotor_foils = (
    Foil('BMSTU', discreteness, 
         chord=0.05, installation_angle=radians(30), step=0.03, name='rotor_1',
         rotation_angle=radians(70),
         relative_inlet_radius=0.05, relative_outlet_radius=0.025,
         inlet_angle=radians(20), outlet_angle=radians(10),
         x_ray_cross=0.4,
         upper_proximity=0.5),
    Foil('BMSTU', discreteness, 
         chord=0.05, installation_angle=radians(30), step=0.03, name='rotor_2',
         rotation_angle=radians(70),
         relative_inlet_radius=0.05, relative_outlet_radius=0.025,
         inlet_angle=radians(20), outlet_angle=radians(10),
         x_ray_cross=0.4,
         upper_proximity=0.5),
    Foil('BMSTU', discreteness, 
         chord=0.05, installation_angle=radians(30), step=0.03, name='rotor_3',
         rotation_angle=radians(70),
         relative_inlet_radius=0.05, relative_outlet_radius=0.025,
         inlet_angle=radians(20), outlet_angle=radians(10),
         x_ray_cross=0.4,
         upper_proximity=0.5),
)
```

```python
figsize: tuple = (15, 15)  # размер изображения в дюймах
```

```python
for foil in (*stator_foils, *rotor_foils):
    fig = foil.plot(figsize=figsize)
    fig.savefig(f'./exports/{foil.name}.png')
    fig.show()
```
